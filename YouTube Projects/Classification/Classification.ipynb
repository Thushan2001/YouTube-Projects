```python
import pandas as pd
import matplotlib.pyplot as plt
import tensorflow as tf
import opendatasets as od
```

```python
od.download("https://www.kaggle.com/datasets/saranpannasuriyaporn/male-female-height-and-weight")
```

```python
data = pd.read_csv("/content/male-female-height-and-weight/Test set.csv")
```

```python
data.shape
```

```python
data.head(5)
```

```python
data.loc[data["Sex"]=="Male","Sex"]=1
data.loc[data["Sex"]=="Female","Sex"]=0
```

```python
data.head(5)
```

```python
x = data.drop(["Sex"],axis = 1)
y = data["Sex"]
y = int(y)
```

```python
x.head(5)
```

```python
y.head(5)
```

```python
plt.scatter(x["Height"],x["Weight"], c = y)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.colorbar()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 42)
```

```python
x_train.shape, x_test.shape
```

```python
data.shape
```

Model Building

```python
#create a model
model = tf.keras.Sequential([
    tf.keras,layers.Dence(4,activation="relu", input_shape=(2,)),
    tf.keras.layers.Dence(1,activation="sigmoid")

])

#compile model
model.compile(loss=tf.keras.losses.BinaryCrossentropy(),
              optimizer = tf.keras.optimizer.Adam(lr = 0.01),
              metrics = ["accuracy"])

#train model
epochs_number = 10
history = model.fit(x_train, y_train, epochs=epochs_number)
```

```python
model.save("new")
```
